```python
#Fixed Acidity histogram 
plt.figure(figsize=(11,6))
sns.histplot(data = data,x = 'fixed acidity', color='blue', edgecolor = 'linen',alpha = 0.5, bins=5)
plt.title("Histogram of fixed acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.show()
```

```python
round(data['fixed acidity'].mean(),2)
```

```python
data['fixed acidity'].median()
```

```python
#Fixed Acidity histogram with mean and median
plt.figure(figsize=(11,6))
sns.histplot(data = data, x = 'fixed acidity', color= 'turquoise', edgecolor = 'purple', alpha = 0.5, bins=5)

plt.title("Histogram of fixed acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.vlines(data['fixed acidity'].mean(),ymin=0,ymax=4000, colors='purple', label='Mean')
plt.vlines(data['fixed acidity'].median(),ymin=0,ymax=4000, colors='blue', label='Median')
plt.legend()
plt.show()
```

```python
#Volatile Acidity histogram
plt.figure(figsize = (11,6))
sns.histplot(data = data, x = 'volatile acidity', color= 'grey', edgecolor = 'black', alpha = 0.5, bins=5)

plt.title("Histogram of volatile acidity")
plt.xlabel('Volatile Acidity')
plt.ylabel('Count')
```

```python
#Distribuition of Volatile Acidity
plt.figure(figsize=(11,6))
sns.distplot(data['volatile acidity'], color = 'blue')


plt.title("Distplot of Fixed Acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('Density')
plt.show()
```
