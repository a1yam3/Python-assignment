```python
#Scatter plot of the age and employee salary
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='age', y='salary',color = 'green',
                edgecolor = 'black', alpha=0.5)
plt.title("Scatterplot of Employee Age and Employee Salary")
plt.xlabel("Employee Age")
plt.ylabel("Employee Salary")
plt.show()
```

```python
#Scatterplot of mployee healthy_eating and active_lifestyle features
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='healthy_eating', y='active_lifestyle',color = 'blue',
                edgecolor = 'black', alpha=0.5)
plt.title("Scatterplot of Healthy Eating and Active Lifestyle")
plt.xlabel("Healthy Eating")
plt.ylabel("Active Lifestyle")
plt.show()
```

```python
#Scatterplot of employee healthy_eating and salary feature
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='healthy_eating', y='salary',color = 'yellow',
                edgecolor = 'black', alpha=0.5)
plt.title("Scatterplot of Healthy Eating and Active Lifestyle")
plt.xlabel("Healthy Eating")
plt.ylabel("Employee Salary")
plt.show()
```

```python
#Countplot of employee groups feature
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='groups', palette=['#FF5733', '#33FF57', '#3357FF'],edgecolor = 'black', alpha=0.7)
plt.title("Countplot of Employee Groups")
plt.xlabel("Employee Groups")
plt.ylabel("Count")
plt.show()
```

```python
#Histogram of the salary feature
plt.figure(figsize = (11,6))
sns.histplot(data = data,x = 'salary', color='blue', edgecolor = 'black',alpha = 0.5, bins=5)
plt.title("Histogram of Employee Salary")
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```
